# Basta Fazoolin'

### Comenzaste un puesto como programador principal del restaurante italiano de estilo familiar Basta Fazoolin' with My Heart . Al restaurante le ha ido fantásticamente y ha visto mucho crecimiento últimamente. Te han contratado para mantener las cosas organizadas.


## Haciendo los Menus


1. En Basta Fazoolin' with my Heart, nuestro lema es simple: cuando estás aquí con la familia, ¡es genial! Disponemos de cuatro menús diferentes: brunch, madrugador, cena y niños.

    Crea una Menu class.


2. Proporcione Menu un constructor con los cinco parámetros self, name, items, start_time y end_time.


3. Vamos a crear nuestro primer menú: brunch. El brunch se sirve de 11 a 16 horas. Los siguientes artículos se venden durante el brunch:

    {
      'pancakes': 7.50, 'waffles': 9.00, 'burger': 11.00, 'home fries': 4.50, 'coffee': 1.50, 'espresso': 3.00, 'tea': 1.00, 'mimosa': 10.50, 'orange juice': 3.50
    }


4. Vamos a crear nuestro segundo elemento de menú early_bird. Las cenas para madrugadores se sirven de 15:00 a 18:00 horas. Los siguientes artículos están disponibles durante el menú de reserva anticipada:

    {
      'salumeria plate': 8.00, 'salad and breadsticks (serves 2, no refills)': 14.00, 'pizza with quattro formaggi': 9.00, 'duck ragu': 17.50, 'mushroom ravioli (vegan)': 13.50, 'coffee': 1.50, 'espresso': 3.00,
    }
    
    
5. Vamos a crear nuestro tercer menú, dinner. La cena se sirve de 17:00 a 23:00 horas. Los siguientes artículos están disponibles para la cena:

    {
      'crostini with eggplant caponata': 13.00, 'caesar salad': 16.00, 'pizza with quattro formaggi': 11.00, 'duck ragu': 19.50, 'mushroom ravioli (vegan)': 13.50, 'coffee': 2.00, 'espresso': 3.00,
    }
    
    
6. Y vamos a crear nuestro último menú, kids. El menú infantil está disponible de 11 a 21 horas. Los siguientes elementos están disponibles en el menú para niños.

    {
      'chicken nuggets': 6.50, 'fusilli with wild mushrooms': 12.00, 'apple juice': 3.00
    }
    
    
7. Dale a nuestra Menu class un método de representación de cadenas que te dirá el name menú. Además, indique en esta representación cuando el menú está disponible.


8. Pruebe nuestra representación de cadenas. Si lo llama print(brunch), debería imprimir algo como lo siguiente:

    brunch menu available from 11am to 4pm
    
    
9. Proporcione Menuun método .calculate_bill()que tenga dos parámetros: self, y purchased_items, una lista de los nombres de los artículos comprados.

    Ha calculate_bill de devolver el precio total de una compra que consta de todos los artículos en purchased_items.


10. Prueba Menu.calculate_bill(). Tenemos una orden de desayuno para una orden de panqueques, una orden de papas fritas caseras y un café. Pase eso brunch.calculate_bill() e imprima el precio.


11. ¿Qué pasa con una compra anticipada? Nuestros últimos invitados pidieron el plato de salumeria y los raviolis veganos de champiñones. Calcula la factura con calculate_bill().


## Creando las Franquicias


12. Basta Fazoolin' with my Heart ha tenido un éxito tremendo en el mercado familiar, lo cual es fantástico porque cuando estás en Basta Fazoolin' with my Heart with family, ¡es fantástico!

    Hemos decidido crear más de un restaurante para ofrecer nuestros fantásticos menús, servicios y ambiente en todo el país.

    Primero, vamos a crear una Franchise class.


13. Dale a la Franchise class un constructor. Tome un address, y asígnelo a self.address. También tome una lista de menus y asígnela a self.menus.


14. ¡Creemos nuestras dos primeras franquicias! Nuestra tienda insignia se encuentra en "1232 West End Road" y nuestra nueva entrega se encuentra en "12 East Mulberry Street". Pase los cuatro menús junto con estas direcciones para definir flagship_store y new_installment.


15. Proporcione nuestra Franchise representación de cadenas para que podamos distinguirlos. Si imprimimos un Franchise nos debe indicar la dirección del restaurante.


16. ¡Digamos a nuestros clientes lo que pueden pedir! Proporcione Franchise un available_menus() método que tome un time parámetro y devuelva una lista de los Menu objetos que están disponibles en ese momento.


17. ¡Probemos nuestro available_menus() método! Llámelo con 12 del mediodía como argumento e imprima los resultados.


18. ¡Hagamos otra prueba! Vea lo que se imprime si llamamos available_menus() con 5pm como argumento e imprima los resultados.


## Creando Negocios!


19. Dado que hemos tenido tanto éxito en la creación de una cadena de restaurantes de marca, hemos decidido diversificarnos. ¡Vamos a crear un restaurante que venda arepas!

    Primero vamos a definir una Business class.


20. Dar a Business un constructor. Business necesita una name y una lista de franchises.


21. Vamos a crear nuestro primero Business. El nombre es "Basta Fazoolin' with my Heart" y las dos franquicias son flagship_store y new_installment.


22. Antes de crear nuestro nuevo negocio, necesitaremos un Franchise y antes de nuestro Franchise necesitaremos un menú. Los artículos para nuestro Take a' Arepa disponibles desde las 10am hasta las 8pm son los siguientes:

    {
      'arepa pabellon': 7.00, 'pernil arepa': 8.50, 'guayanes arepa': 8.00, 'jamon arepa': 7.50
    }
    
    Guarde esto en una variable llamada arepas_menu.


23. A continuación, ¡creemos nuestra primera franquicia Take a' Arepa! Nuestro nuevo restaurante está ubicado en "189 Fitzgerald Avenue". Guarde el Franchise objeto en una variable llamada arepas_place.


24. ¡Ahora hagamos nuestro nuevo Business! El negocio se llama "Take a' Arepa"!


25. ¡Felicitaciones! Creó un sistema de clases que ayudan a estructurar su código y realizar todos los requisitos comerciales que necesita. Cada vez que necesitemos una nueva función, tendremos el código bien organizado necesario para facilitar su desarrollo y envío.
